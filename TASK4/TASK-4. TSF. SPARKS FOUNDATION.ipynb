# Exploratory Data Analysis


## Countries that are most affected  and least affected by terrorist attacks

```python
df['Country'].value_counts().head()    ## Countries that are most affected by terrorist attacks
```

```python
df['Country'].value_counts().tail()   ## Countries that are least affected by terrorist attacks
```

```python
print("Country with the most attacks:", df['Country'].value_counts().idxmax())   ## Country with maximum number of terrorist attacks
```

```python
df['Country'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=df['Country'].value_counts()[:15].index,y=df['Country'].value_counts().values[:15], palette = 'rocket')
plt.title('Countries that got attacked the most')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Attacks')
plt.show()
```

```python
#Visualising the countries affected by terrorist attacks with a  pie chart
df['Country'].value_counts().head() 
df['Country'].value_counts().head().plot(kind = 'pie', figsize = (20,8), fontsize=10, autopct= '%3.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Countries", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)



```

```python
df['Country'].value_counts().tail() 
df['Country'].value_counts().tail().plot(kind = 'pie', figsize = (20,8), fontsize=10, autopct= '%3.1f%%', 
                                                shadow=True, startangle=135, legend=True, )
plt.ylabel("Countries", size = 10)
plt.xlabel("Attacks Occured", size = 10)
plt.xticks(size = 15)
plt.yticks(size = 15)
```


# Observation :  

**1.Iraq has the maximum no of terrorist attacks with number near about 25000 followed by Pakistan , Afghanistan and then India.**

**2.Countries like South Vietnam , St. Lucia , Andorra, Falkland Islands are less affected by terrorist attacks.**

# Visualising the trend of deaths occured and wounded people over the years

```python
## Visualization the trend of deaths with the help of a line graph


df.groupby('Year')['Killed'].sum().plot(kind = 'line', linewidth = 5, figsize =(15,5),color = 'red')
plt.title("Trend of deaths due to terror attacks",fontsize= 10, fontweight = 'bold')
plt.xlabel("Years", size = 16)
plt.ylabel("Total deaths occured ", size = 10)
plt.xticks(size = 10)
plt.yticks(size = 10)
```

```python
##Visualising the trend of wounded people over the years with the help of a line graph

df.groupby('Year')['Wounded'].sum().plot(kind = 'line', linewidth = 5, figsize =(15,5),
                                                                                             color = 'red')
plt.title("People wounded due to terror attacks",fontsize= 10, fontweight = 'bold')
plt.xlabel("Years", size = 16)
plt.ylabel("Total wounded people ", size = 10)
plt.xticks(size = 10)
plt.yticks(size = 10)
```

# Observations:

**1. There has been  a  steady increase in the no. of deaths as well as injured people .In the years 1990-2010 there has been fluctuations.**


**2. From the year 2010 onwards, the no. of deaths has reached its highest peak with the increase in the number of terror attacks.**
